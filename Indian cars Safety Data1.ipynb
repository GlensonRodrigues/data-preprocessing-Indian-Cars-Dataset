Diving between Featues & target for our model

```python
features=df.iloc[:,:-1]
target=df.iloc[:,-1]
```

```python
features.columns
```

```python
df["Fuel Efficiency"].fillna(df["Fuel Efficiency"].median(),inplace=True)
```

```python
columns=features.select_dtypes(["object"]).columns
columns
```

```python
features.info()
```

Converting from Object to float dtype

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
oe.fit_transform(features[["Type","Fuel"]])
```

```python
features[columns]=oe.fit_transform(features[columns])
```

successfully chnage the dtypes to Numbers & now running model.

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(features,target,test_size=0.3,random_state=1)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
hbc = HistGradientBoostingClassifier()
hbc.fit(xtrain,ytrain)
ypred=hbc.predict(xtest)

from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(ytest,ypred)
rmse=np.sqrt(mse)
r2=r2_score(ytest,ypred)
print(mse)
print(rmse)
print(r2)
```
